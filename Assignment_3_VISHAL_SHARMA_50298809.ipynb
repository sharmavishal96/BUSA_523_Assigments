Question 1: Create a class to store student records. Each student has a first name, last name,
country, overall grade (0 - 4.0 scale), credits completed (total 30) and major.
Create a method within this class to evaluate student progress.

```python
class student_records:
    def input_data(self, Fname, Lname, country, overall_grade, credits_completed, smajor):
        student_records.Fname = Fname
        student_records.Lname = Lname
        student_records.country = country
        student_records.overall_grade = overall_grade
        student_records.credits_completed = credits_completed
        student_records.smajor = smajor
        
   
    def evaluate_student_progress(self):
        print("\nStudent Details is given as: ")
        print("Name: ", student_records.Fname, student_records.Lname)
        print("Country: ", student_records.country)
        print("overall grade: ", student_records.overall_grade)
        print("credits completed: ", student_records.credits_completed)
        print("Major: ", student_records.smajor)
        print("\n")
        
        if student_records.overall_grade >= 3.0 and student_records.credits_completed >= 20:
            print("Status: GOOD")
        elif (student_records.overall_grade >= 2.5 and student_records.overall_grade < 3.0) or (student_records.credits_completed >=10 and student_records.credits_completed <20):
            print("Status: AVERAGE")
        elif student_records.overall_grade < 2.5:
            print("Status: BAD")

Fname = input("Enter first name: ")
Lname = input("Enter last name: ")
country = input("Enter Country: ")
overall_grade = float(input("Enter overall grade(0 - 4.0 scale): "))
credits_completed = int(input("Enter credits completed(total 30): "))
smajor = input("Enter major: ")
            
s1 = student_records()
s1.input_data(Fname, Lname, country, overall_grade, credits_completed, smajor)

s1.evaluate_student_progress()
            
 
```

Question 2: Write a function which separates the positive and negative numbers from a
list of numbers

```python
input_list = []    
num =  int(input("Enter total number of input you want in the list"))
for j in range (1, num +1):
    num_list = int(input("Enter number "))
    input_list.append(num_list)

def sep_num():
        pos_list = []
        neg_list = []

        for i in range (len(input_list)):
            if input_list[i] >0:
                pos_list.append(input_list[i])
            elif input_list[i] <0:
                neg_list.append(input_list[i])
        print("Output: ", neg_list,",", pos_list)
    
sep_num()
```

Question 3: Write a function which outputs all winning possibilities for a given playerâ€™s
score. Assume the score is always between 15 and 21. Further, the dealer is
dealt only 2 cards.

```python
number = int(input("Enter your input(between 15 and 21): "))

def black_jack():
    combinations = []
    output = []
    for i in range (1,12):
        for j in range (i,12):
            if number < i + j < 22 :
                combinations.append(i)
                combinations.append(j)
                output.append(combinations)
                combinations = []
    print(output)
        
black_jack()
```
